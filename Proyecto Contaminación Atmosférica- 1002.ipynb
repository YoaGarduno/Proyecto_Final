#### Para calcular nuestro modelo, utilizaremos el método de mínimos cuadrados con las siguientes fórmulas:

$$ y = mx + b $$

$$ 
m = \frac{\sum x \sum y - n\sum(xy)}{\left(\sum x\right)^2 - n\sum x^2}
$$
y|
$$
b = \overline{y} - m\overline{x}
$$

$$
R = \frac{\sigma_{xy}}{\sigma_x \sigma_y}
$$
donde
$$
\sigma_x = \sqrt{\frac{\sum\left(x^2\right)}{n} - \overline{x}^2}, \quad
\sigma_y = \sqrt{\frac{\sum\left(y^2\right)}{n} - \overline{y}^2}
$$
$$
\sigma_{xy} = \frac{\sum(xy)} - \overline{x}\cdot\overline{y}
$$

Primero, para CEO3:

```python
x = [2010., 2011., 2012., 2013., 2014., 2015., 2016., 2017., 2018., 2019., 2020.]   #A  x le asignamos los valores del arreglo correspondiente a los años
y = [75.052055, 77.854795, 73.079235, 76.30411 , 73.205479, 76.439891, 83.520436, 86.428962, 85.42623 , 81.062842, 79.817439]   #De igual forma para y con los areglos de CEO3
```

```python
n = len(y)         #Le asignamos a n el valor de el número de datos que contiene x o y, los cuales, deben tener la misma cantidad 
x = np.array(x)     #Asignamos los arreglos a las variables
y = np.array(y)
x           #Indicamos que se muestre x para confirmar que no haya un error
```

```python
sumx = sum(x)     #En esta parte, asignamos los valores que van a tomar las variables que necesitamos en la ecuación, la suma de x
sumy = sum(y)      #La suma de y
sumx2 = sum(x*x)   #La suma de los cuadrados en x
sumy2 = sum(y*y)   #La suma de los cuadrados en y
sumxy = sum(x*y)   #La suma de multiplicar los valores de x por los de y
promx = sumx/n     #El promedio de x
promy = sumy/n     #El promedio de y
```

```python
m = (sumx*sumy - n*sumxy)/(sumx**2 - n*sumx2)  #Utilizamos las variables que creamos anteriormente para el desarrollo de las fórmulas para la pendiente (m)
b = promy - m*promx         #Y para la ordenada al origen (b)
m , b                   #Indicamos que se muestren los valores resultantesde m y b
```

```python
plt.plot(x, y, 'o', label='Datos')         #Graficamos nuestro modelo de mínimos cuadrados 
plt.plot(x, m*x + b, label='Ajuste')
plt.xlabel('Año')
plt.ylabel('CEO3')
plt.title('Índice de Calidad del Aire, Centro Ciudad de México, CEO3')
plt.grid()
plt.legend()
plt.show()
```

```python
sigmax =  np.sqrt(sumx2/n - promx**2)     #En esta celda, designamos los valores necesarios para conocer nuestro coeficiente de correlación, que nos dice qué tan preciso es nuestro ajuste
sigmay =  np.sqrt(sumy2/n - promy**2)
sigmaxy = sumxy/n - promx*promy
R2 = (sigmaxy/(sigmax*sigmay))**2
R2       #Nuestro coeficiente es de 0.44, el cual podría aumentar si ampliaramos nuestra muestra
```

```python
def my_CEO3(t):           #Definimos una función que nos permita conocer cuál será el valor promedio de la calidad del aire con el Ozono para t año
    t >=2000
    o = m* t + b

    return o
```
